```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
sns.set(color_codes = True)
```

```python
sns.boxplot(data = data, y = 'Avg_Open_To_Buy')
plt.title('boxplot for the Avg_Open_To_Buy column')
plt.show
```

```python
sns.boxplot(data = data, y = 'Total_Trans_Ct')
plt.title('boxplot for the Total_Trans_Ct column')
plt.show
```

```python
pie_chart = data['Gender'].value_counts()
plt.pie(pie_chart, labels=pie_chart.index,autopct='%1.1f%%')
plt.axis('equal')
plt.title('Pie chart of the Gender')
plt.show
```

```python
pie_chart2 = data['Education_Level'].value_counts()
plt.pie(pie_chart2, labels=pie_chart2.index,autopct='%1.1f%%')
plt.axis('equal')
plt.title('Pie chart of the Education_Level')
plt.show
```

```python
box_colors = ['red', 'blue', 'green']
whisker_color = 'black'

sns.boxplot(data = data, x = 'Total_Revolving_Bal', y = 'Card_Category', hue = 'Attrition_Flag', 
           palette = box_colors, boxprops = {'edgecolor' : whisker_color})

plt.xlabel('Total Revolving Bal')
plt.ylabel('Card Category')
plt.title('Box plot of Total Revolving Balance by Card category and Attrition Flag')

plt.legend(title = 'Attrition Flag')
plt.show
```

```python
percentage_graph1 = data.groupby('Education_Level')['Attrition_Flag'].value_counts(normalize = True).unstack()
percentage_graph1.plot(kind = 'bar', stacked = True)
plt.title('Percentage Segment Bar Graph Between Education_Level and Attrition_Flag of the customers')
plt.xlabel('Education_Level')
plt.ylabel('Percentage')
plt.legend(title = 'Attrition_Flag', loc = 'upper right')
plt.show
```

```python
percentage_graph2 = data.groupby('Income_Category')['Attrition_Flag'].value_counts(normalize = True).unstack()
percentage_graph2.plot(kind = 'bar', stacked = True)
plt.title('Percentage Segment Bar Graph Between Income_Category and Attrition_Flag of the customers')
plt.xlabel('Income_Category')
plt.ylabel('Percentage')
plt.legend(title = 'Attrition_Flag', loc = 'upper right')
plt.show
```
